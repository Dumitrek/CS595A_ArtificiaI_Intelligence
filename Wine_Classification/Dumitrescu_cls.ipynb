Importing the Wine dataset along with all the necessary libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from pandas.plotting import scatter_matrix
%matplotlib notebook

wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header = None)
```

Outputting the dataset as a whole with properly assigned column headings

```python
wine.columns = ['Classes', 'Alcohol', 'Malic Acid', 'Ash', 'Alcalinity of Ash', 'Magnesium', 'Total phenols', 'Flavanoids', 
                'Nonflavoids phenols', 'Proanthocyanins', 'Color Intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
wine
```

Outputting the shape on the wine dataset via the shape function call on the dataset

```python
wine.shape
```

Displaying the min, max, and average values for the feature names and target labels via the use of the describe function

```python
wine.describe()
```

Displaying the scatter matrix for the given dataset using the scatter_matrix function call with a configured and adjustable plot window

```python
scatter_matrix(wine)
fig = plt.figure(figsize=(10000,10000), dpi=200, facecolor='w', edgecolor='k')
plt.show()
```

Training the dataset first by spliting the data

```python
variables = wine.values
x = variables[:, 1:13]
y = variables[:, 0]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

Preprocessing the dataset

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

K-Nearest Neighbors Implementation

```python
perceptor = KNeighborsClassifier(n_neighbors=3)
perceptor.fit(x_train, y_train)
y_pred =perceptor.predict(x_test)
print("Confusion Matrix: \n" + str(confusion_matrix(y_test, y_pred)))
print("\nAccuracy score: \n" + str(accuracy_score(y_test, y_pred)))
print("\nPerformance Statistics: \n" + str(classification_report(y_test, y_pred)))
plt.show(perceptor)
```

Logistic Regression Implementation

```python
perceptor = LogisticRegression(C=100, max_iter=1500, random_state=1)
perceptor.fit(x_train, y_train)
y_pred =perceptor.predict(x_test)
print("Confusion Matrix: \n" + str(confusion_matrix(y_test, y_pred)))
print("\nAccuracy score: \n" + str(accuracy_score(y_test, y_pred)))
print("\nPerformance statstics: \n" + str(classification_report(y_test, y_pred)))
```
